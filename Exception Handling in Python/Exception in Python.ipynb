**Exceptions**

_If something goes wrong while the program is running (runtime), it is called an Exception._

1. It usually happens when something unexpected occurs.

2. Exceptions are raised by the Python runtime.

3. You must handle them at that moment (while the program is running).

**Examples of Exceptions**

1. _Memory overflow_

2. _Dividing by 0 (logical error)_

3. _Database error_

```python
# catching specific exception
try:
 m = 5
 f = open('text.txt','r')
 print(f.read())
 print(m)
 print(5/3)
 l = [1,2,3,4, 5]
 l[100]
except FileNotFoundError:
    print('file not found Error !')
except NameError:
    print('Variable not defined Error !')
#except Exception as e:
    #print(e.with_traceback)
except ZeroDivisionError:
    print('division by zero Error !')
except Exception as e:
    print(e)
```

```python
# else

try :
    f = open('text.txt','r')
except FileNotFoundError:
    print('file not found Error !')
except Exception:
    print("Kuch toh Gadbad hai daya !")
else:
    print(f.read())
```

```python
# finally

try :
    f = open('text.txt','r')
except FileNotFoundError:
    print('file not found Error !')
except Exception:
    print("Kuch toh Gadbad hai daya !")
else:
    print(f.read())
finally:
    print('This will definitely gonna Execute ! ')
```

```python
# Example for raise :

class bank:
    def __init__(self,balance):
        self.balance = balance

    def withdraw(self,amount):
        if amount < 0:
            raise Exception('Balance cannot be negative')
        if self.balance < amount:
            raise Exception('not enough money')
        self.balance -= amount

obj = bank(1000)
try:
 obj.withdraw(-500)
except Exception as e:
    print(e)
else:
    print(obj.balance)
```
