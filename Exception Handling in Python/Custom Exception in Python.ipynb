```python
# creating custom Exceptions
# exception hierarchy in python.


class MyException(Exception):   # our own custom Exception !
    def __init__(self, message):
        print(message)


class Bank:
    def __init__(self, amount):
        self.balance = amount
class bank:
    def __init__(self,balance):
        self.balance = balance

    def withdraw(self,amount):
        if amount < 0:
            raise MyException('Balance cannot be negative')
        if self.balance < amount:
            raise MyException('not enough money')
        self.balance -= amount

obj = bank(1000)
try:
 obj.withdraw(-500)
except MyException as e:
    print(e)
else:
    print(obj.balance)
```

```python
# example for Custom Exception :

# Google Example :

class SecurityError(Exception):
    def __init__(self, message):
        print(message)

    def logout(self):
        print('Logout Successful')


class google:

    def __init__(self, name, email, password, device):
        self.name = name
        self.email = email
        self.password = password
        self.device = device
    def login(self, email, password, device):
        if device != self.device:
            raise SecurityError('bro sorry for that hehehhehehe')
        if email == self.email and password == self.password:
            print('Login Successful')
        else :
            print('Login Failed')

if __name__ == '__main__':
    obj = google(name='ayush suthar', email='ayushsuthar@gmail.com', password='1234', device='Android')
    try:
        obj.login('ayushsuthar@gmail.com', '1234', 'Apple') # device changed
    except SecurityError as e :
        e.logout()
    else:
        print(obj.name)
    finally:
        print('database connection closed !')
```
