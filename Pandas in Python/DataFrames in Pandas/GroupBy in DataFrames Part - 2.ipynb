```python
import pandas as pd
import numpy as np

movies = pd.read_csv('/workspaces/Python-for-ML/Database from Kaggle/imdb-top-1000.csv')
genres = movies.groupby('Genre')
```

```python
# agg method
# passing dict
genres.agg({
    'Runtime': 'mean',
    'IMDB_Rating': 'mean',
    'No_of_Votes': 'sum',
    'Gross': 'sum',
    'Metascore': 'min'
})
```

```python
# adding both the syntax 
genres.agg({
    'Runtime': ['mean','min'],
    'IMDB_Rating': ['mean','max'],
    'No_of_Votes': ['sum', 'max'],
    'Gross': 'sum',
    'Metascore': 'min'
})
```

```python
# passing list
genres[['Runtime', 'IMDB_Rating', 'No_of_Votes', 'Gross', 'Metascore']].agg(['min', 'max', 'mean'])
```

```python
# Looping on groups

for group,data in genres :
    print(type(group), type(data))
```

```python

df = pd.DataFrame(columns=movies.columns)
print(df)
for group, data in genres:
    top_rated = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df = pd.concat([df, top_rated], ignore_index=True)

df
```

```python
# split ( apply ) combine
# apply -> builtin function

genres.apply(min)
```

```python
# find number of movies starting with A for each group

def ayush(group):
    return group['Series_Title'].str.startswith('A').sum()
genres.apply(ayush)
```

```python
# other method : 

genres.apply(lambda group: group['Series_Title'].str.startswith('A').sum())
```

```python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
    group['Genre_Rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group


genres.apply(rank_movie)
```

```python
# Find Normalized IMDB rating group wise 

def normal(group):
    group['norm_rating'] = group['IMDB_Rating'] - group['IMDB_Rating'].min()/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group



genres.apply(normal)


```
