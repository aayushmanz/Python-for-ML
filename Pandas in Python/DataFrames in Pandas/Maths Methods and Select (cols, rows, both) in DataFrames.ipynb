**1. Math Methods**

```python
# Sum -> axis arguements

import pandas as pd
import numpy as np

student_dict = {
    'iq' : [80,70,120,80,0,0],
    'marks' : [100,70,100,50,0,0],
    'package' : [10,7,14,2,0,0]
}
std_dict = pd.DataFrame(student_dict)
std_dict.sum()
```

```python
# Sum :
std_dict.sum(axis=1)
```

```python
# mean, median, mode
std_dict.mean(axis=1)
```

```python
std_dict.mode(axis=1)
```

```python
std_dict.median(axis=1)
```

```python
# var
std_dict.var(axis=True)
```

**3. Selecting rows in a dataframes :**


*   iloc -> searches using index positions
*   loc -> searches using index labels



1. iloc()

```python
# single row :
movies =  pd.read_csv('/workspaces/Python-for-ML/Database from Kaggle/movies.csv')
movies.iloc[0]  # iloc use numbers for getting the data from the data base.
# i use the 0 as a index positon for getting the first data 
```

```python
# multiple row :
movies.iloc[0:5] # now i get 0 to 5 movies on index 
```

```python
movies.iloc[5:16:2]
```

```python
movies.iloc[0:3,0:3]
```

```python
# fancy index :

movies.iloc[[0,4,5]]
```
