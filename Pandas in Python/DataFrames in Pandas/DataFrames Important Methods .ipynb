```python
students = pd.DataFrame(
    {
        'name': ['ritish', 'ankit', 'rupesh', np.nan, 'sunitanjay', np.nan, 'nishabh', np.nan, 'aditya', np.nan],
        'college': ['bit', 'iit', 'vit', np.nan, np.nan, 'vlsi', 'ssit', np.nan, np.nan, 'sit'],
        'branch': ['eee', 'it', 'cse', np.nan, np.nan, 'cse', 'bio', np.nan, np.nan, np.nan],
        'cgpa': [6.66, 8.25, 6.41, np.nan, 5.6, 9.0, 7.4, 10, 7.4, np.nan],
        'package': [4, 5, 6, np.nan, 6, 7, 8, 9, np.nan, np.nan]
    }
)

students
```

```python
students.sort_values('name', na_position='first', ascending=False,) 
# name column sorted and pushed the NaN values at the backside of the ranking.
```

```python
students
```

```python
movies.head(1)
movies.sort_values(['year_of_release', 'title_x'], ascending=[True,False])
# we can add multiple cols by using the list in sort values and we can also use ascending for mutiple cols by list.
```
