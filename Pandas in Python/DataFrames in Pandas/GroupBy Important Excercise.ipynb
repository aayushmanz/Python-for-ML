```python
import pandas as pd
import numpy as np

ipl = pd.read_csv('/workspaces/Python-for-ML/Database from Kaggle/deliveries.csv')
ipl.head()
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs scores.

ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
# find which batsman with most no of sixes.

six = ipl[ipl['batsman_runs']==6]

six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find batsman with most number of 4 and 6 in last overs

temp = ipl[ipl["over"]>15]
temp=temp[(temp['batsman_runs']==4) | (temp['batsman_runs']==6)]
temp.groupby('batsman')['batsman'].count().sort_values(ascending=False).head()
```

```python
# find V kohli's record against all terms 
temp = ipl[ipl['batsman']=='V Kohli']
temp.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can returns the highest score of any batsman

temp = ipl[ipl['batsman'] == 'V Kohli']
temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp = ipl[ipl['batsman'] == batsman]
  return temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('DA Warner')
```
