**What is Pandas**

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

**Pandas Series**

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

# importing Pandas :

```python
import numpy as np
import pandas as pd
```

**Series From lists :**

```python
# string :

country = ['india', 'pakistan', 'usa', 'japan']

pd.Series(country) # using series in the list of strings
```

```python
# integer :

runs = [100, 56, 45, 36, 27]

runs_ser = pd.Series(runs)  # using series in the list of numbers 

runs_ser
```

```python
# custom  index :

marks = [56,34, 78 , 89,23]
subject = ['maths','science','chemistry','chemistry','chemistry' ]

pd.Series(marks, index=subject) # we can also custom index by the use of index=''
```

```python
# setting a name :

marks = pd.Series(marks, index=subject, name='Ayush Marks')  # we can set the name through name=''

marks
```

**Series from dict :**

```python
marks = {
    'maths' : 56,
    'science' : 45,
    'chemistry' : 78,
    'english' : 89
}

marks_series = pd.Series(marks, name='Ayush Marks') # we can also add the name in the dictionaries.

marks_series
```
