**Series Indexing :-**

```python
import pandas as pd
import numpy as np
```

```python
# integer Indexing :
x = pd.Series([12,13,14,35,46,57,59,78,9])
x[1]
```

```python
# negative indexing

x.iloc[-2]
```

```python
# indexing through numbers
movies = pd.read_csv('/content/bollywood.csv')

movies[0:3]
```

```python
movies[1:5:2]
```

```python
movies = pd.read_csv('/content/bollywood.csv',index_col=['lead'])[0:3]
movies
```

```python

movies.loc['Vicky Ahuja'] # we can also find the data through the charaters iloc()
```

```python
movies.iloc[0:1] # we can also use iloc() for searching the index
```

```python
movies.iloc[-1]
```

**Slicing :**

```python
# slicing :

vk = pd.read_csv('/content/kohli_ipl.csv')

vk[5:16]
```

```python
# negitive slicing :

vk[-5:]
```

```python
movies[-5:]
```

```python
movies[::5]
```

```python
# fancy indexing :

vk.iloc[[1,3,4,5]]
```

```python
# indexing with labels -> fancy indexing
movies.loc['Vicky Ahuja']
```
