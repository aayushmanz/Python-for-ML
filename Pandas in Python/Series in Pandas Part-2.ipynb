**2. Series of Methods :**

```python
# Head and tail

matches.head()  # Head takes default -> 5 arrays from the starting rows.
```

```python
matches.head(10) # we can customize our rows.
```

```python
# tails :
matches.tail()  # Tail takes default -> 5 arrays from the ending rows.
```

```python
matches.tail(10) # we can customize our rows.
```
