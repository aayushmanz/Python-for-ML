```python
# What is the average Final_Score and Previous_GPA of all the students in this dataset?

std_data = pd.read_csv("/workspaces/Python-for-ML/Database from Kaggle/student_performance_finalscore.csv")
std_data.sample(3)
```

```python
avg_previous = std_data['Previous_GPA'].mean()
avg_final = std_data['Final_Score'].mean()
print("The Final Year Average : " , avg_final)
print(f'The Previous Year Average : {avg_previous}')
```

```python
std_data.groupby("Age")['Age'].count().sort_values(ascending=False)
```

```python
# Out of the available Study_Method options (Online, Offline, Hybrid), which one is most preferred by the students?
data = std_data.groupby("Study_Method")['Study_Method'].count().sort_values(ascending=False)
print("There are most of studnet Enrolled in : ",data.index[0])
print(f'Total Number of Studnets with {data.index[0]} is {data.iloc[0]}')
```

```python
# Does an increase in Hours_Studied actually result in a higher Final_Score?
std_data.sample(6)
```

```python
corrilation = std_data['Hours_Studied'].corr(std_data['Final_Score'])
print(corrilation.astype('float32'))
```

```python
# Do students who maintain more than 80% Attendance score significantly better than the rest of the class?
high = std_data[std_data["Attendance"]>80.0]['Final_Score'].mean()
low =std_data[std_data['Final_Score']<=80.0]['Final_Score'].mean()

print(f'The High attendance std score : {high:.2f}')
print(f'The Lower attendance std score : {low:.2f}')
```

```python
# Is there a direct link between Exam_Anxiety_Score and Stress_Level? Do students with high stress end up getting lower marks?
a = std_data['Exam_Anxiety_Score'].corr(std_data['Final_Score'])
b = std_data['Stress_Level'].corr(std_data['Final_Score'])

print(f"The Answer for Exam_Anxiety_Score Student final Studnet : {a:.2f} ")
print(f"The Answer for Stress_Level Student final Studnet : {b:.2f} ")
```
