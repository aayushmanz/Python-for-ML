**1. Math Methods**

```python
# Sum -> axis arguements

import pandas as pd
import numpy as np

student_dict = {
    'iq' : [80,70,120,80,0,0],
    'marks' : [100,70,100,50,0,0],
    'package' : [10,7,14,2,0,0]
}
std_dict = pd.DataFrame(student_dict)
std_dict.sum()
```

```python
# Sum :
std_dict.sum(axis=1)
```

```python
# mean, median, mode
std_dict.mean(axis=1)
```

```python
std_dict.mode(axis=1)
```

```python
std_dict.median(axis=1)
```

```python
# var
std_dict.var(axis=True)
```
