```python
subs = pd.read_csv('/workspaces/Python-for-ML/Database from Kaggle/subs.csv')
subs
```

```python
subs.mean()
```

```python
# It will Squeeze the DataFrame and convert it into Series.
subs = subs.squeeze()
```

```python
data = subs.apply(lambda x : 'Good Day' if x > subs.mean() else 'bad day')
Gcount = 0
Bcount = 0

for i in data:
 if i == 'Good Day':
    Gcount = Gcount + 1
 else :
    Bcount = Bcount + 1    
print(f'Our Total good days of subs is {Gcount}')
print(f'Our bad count is {Bcount}')
```
