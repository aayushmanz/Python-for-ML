# Editing Series :

```python
import pandas as pd
import numpy as np

marks_series = {
    "English" : 67,
    "Hindi"   : 78,
    "Science" : 89,
    "Maths"   : 90
}

marks = pd.Series(marks_series, name="Ayush Suthar")
print(marks)

```

```python
# using indexing :
marks[1] = 100
print(marks)
```

```python
marks['Hindi'] = 99
print(marks)
```

```python
marks[1:1] = 99
print(marks)
```

```python
# what if an index does not exist
marks['Social'] = 78

marks
```

```python
# slicing :

runs = pd.read_csv("/content/kohli_ipl.csv").head()['runs']

runs
```

```python
runs[1:3] = [100,100]

runs
```

```python
# fancy index label
runs[[0,3,4]] = [0,0,0]

runs
```

```python
# using index label

movies = pd.read_csv("/content/bollywood.csv").head()

movies
```

```python
movies.loc['Why Cheat India'] = 'Ayush Suthar'
movies
```

**4. looping**

```python
movies = pd.read_csv("/content/bollywood.csv").head(10)

movies
```

```python
for i in movies : # for columns
  print(i)
```

```python
for i in movies.index :  # for index
  print(i)
```

```python
for i in movies.values : # for values
  print(i)
```

```python
for i in movies['movie'] : # for the index values
  print(i)
```

```python
for i in movies['lead'] : # for index values
  print(i)
```

# Boolean Indexing on Series :

```python
vk = pd.read_csv("/content/kohli_ipl.csv")

vk
```

```python
# find no of 50's and 100's scored by kohli :

vk[vk['runs'] >= 30].size
```

```python
# find number of ducks

vk[vk['runs']==0]
```

```python
vk[vk['runs']==0].size
```

```python
# count number of day when I had more than 200 subs a day
subs = pd.read_csv("/content/subs.csv")

subs[subs['Subscribers gained'] > 200]
```

```python
subs[subs['Subscribers gained'] > 200].size
```

```python
# find actors who have done more than 20 movies.

movies = pd.read_csv("/content/bollywood.csv")

num_movie = movies['lead'].value_counts()
print(num_movie[num_movie > 20])
```
