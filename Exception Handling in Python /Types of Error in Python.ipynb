**A program can have errors at two stages:**

1. _During compilation_ → Syntax Error

2. _During execution_ → Exception

**What is a Syntax Error?**

1. _It happens when the program is not written according to the correct grammar (rules) of the language._

 2. The compiler or interpreter shows this error.

3. You can fix it by correcting the code.

```python
# example of syntax error
print 'ayush bhai'   # if we dont use parentheses
```

**Other examples of Syntax error**
1. leaving symbols like semi colon.
2. missspelling a keyword
3. incorrect indentation
4. empty if/else/loops/class/function
