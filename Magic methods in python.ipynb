**Magic methods** _allow your custom classes (like your Atm class) to behave like Python's built-in data types (like integers, strings, or lists). They let you define how your class should react to standard Python operations._

**1. __init__(self)** (The Initializer)

What triggers it: Creating a new object (e.g., sbi_atm = Atm()).

What it does: It sets up the initial state of your object. As you saw in your code, the moment you create the ATM, __init__ runs automatically, sets the balance to 0, and launches the menu. You never had to write sbi_atm.__init__().

**2. __str__(self)** (The String Representation)

What triggers it: Using the print() function on your object.

What it does: It tells Python how to describe your object in readable text.

Example: Right now, if you type print(sbi_atm) at the bottom of your code, Python will print something ugly like <__main__.Atm object at 0x7f8b...>. But if you define __str__, you can make it print something nice!
