**Serialization and Deserialization**

1. **Serialization** – _process of converting Python data types into JSON format._

2. **Deserialization** – _process of converting JSON data into Python data types._

**What is Json ?**

```python
# serialization using json module
# list
import json

from IPython.utils.text import indent

L = [1,2,3,4]

with open('demo.json', 'w') as f:
    json.dump(L, f)
```

```python
# dict
import json
d = {
    'name' : 'ayush suthar',
    'age' : 22,
    'gender' : 'male',
    'address' : 'Balrampur'
}
with open('demo.json', 'w') as f:
    json.dump(d, f, indent=4)
```

```python
# deserialization

import json

with open('demo.json', 'r') as f:
    print(json.load(f))
```

```python
# serialize and deserialize tuple
import json
t = (1,2,3,4,5)

with open('demo.json', 'w') as f:
    json.dump(t,f)
```

```python
# serialize and deserialize a nested dict

d = {
    'name' : 'ayush suthar',
    'marks' : [75, 65, 55]
}

with open('demo.json', 'w') as f:
    json.dump(d, f, indent=4)
```
