**Problems with working in text mode**

1. _cant work with binary file like images
2. not good for other data types like int / float/ list/ tuples etc_


```python
# working with binary files !
with open('achabacha.jpeg', 'rb') as f: # rb -> read binary !
    with open('pyarabacha.jpeg', 'wb') as wf: # wb -> write binary !
        wf.write(f.read())
```

```python
# working with other data types :
with open('ayush.txt','w') as f:
    f.write(5) # we can not write other data type in write and 5 is int
```

```python
with open('ayush.txt','w') as f:
    f.write('5')
```

```python
with open('ayush.txt','r') as f:
    print(f.read() + 5) # string does not work with string !
```

```python
with open('ayush.txt','r') as f:
    print(int(f.read()) + 5) # type casting !
```

```python
# more complex data :
d= {
    'name' : 'ayush',
    'age' : 22,
    'gender' : 'male'
}

with open('ayush.txt','w') as f:
    f.write(str(d))
```

```python
with open('ayush.txt','r') as f:
    print(f.read())
```
