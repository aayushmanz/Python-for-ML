**Pickling**

_Pickling is the process whereby a Python object hierarchy is converted into a byte stream and unpickling is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy._

```python
class person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def display_info(self):
        print('Hi my name is '+ self.name, 'And I am ' + self.age, 'years old')
```

**Pickle Vs Json**

1. _Pickle lets the user to store data in binary format. JSON lets the user store data in a human-readable text format._

```python
# pickle :

import pickle

data = {"name": "Ayu", "age": 20}

# Serialization (Pickling)
with open("data.pkl", "wb") as f:
    pickle.dump(data, f)

# Deserialization (Unpickling)
with open("data.pkl", "rb") as f:
    new_data = pickle.load(f)

print(new_data)
```

```python
# json :

import json

data = {"name": "Ayu", "age": 20}

# Serialization
with open("data.json", "w") as f:
    json.dump(data, f)

# Deserialization
with open("data.json", "r") as f:
    new_data = json.load(f)

print(new_data)
```
