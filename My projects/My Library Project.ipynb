```python

# I use Encapsulation and static variables in this code.
class library:
    college = "vision college of commerce"  # these are static variables.
    __Total_books = 100
    def __init__(self):
        print('Create Your account : ')
        user = input("Enter Your new username :")
        fname = input("Enter Your first name : ")
        lname = input('Enter your last name : ')
        pasw = int(input("Enter Your new password"))
        self.__username = user              # they are private variables
        self.__password = pasw
        self.fname = fname
        self.lname = lname
        self.book = 0
        self.user_profile ={
            'Username' : self.__username,
            'name' : fname +" "+ lname,
            'College' : library.college,
            'Borrowed Books' : self.book
        }
        self.menu()

    def menu(self):
        choice = input(
            """
            1. borrowing books
            2. returning books
            3. Remaining books
            4. profile
            5. Exit
            """
        )
        if choice == '1':
            self.borrow()
        elif choice == '2':
            self.returning()
        elif choice == '3':
            self.remaining()
        elif choice == '4':
            self.profile()
        elif choice == '5':
            exit()
        else :
            print('Choose the number from 1 to 3 ')
            self.menu()

    def borrow(self):
        username = input('Enter your username : ')
        password = int(input('Enter your password : '))
        print(f'Welcome {self.fname} {self.lname} Sir Thanks for Choosing our services !')
        if password == self.__password and username == self.__username and self.book == 0:
            book = int(input('How many book you want : '))
            self.book = book
            library.__Total_books = library.__Total_books - self.book
            print('Your Book purchase is successful !')
            print(f'You Purchase total {self.book} ')
            self.menu()
        elif password == self.__password and username == self.__username and self.book > 0 :
            print('please kindly return the old books first ! ')
            self.menu()
        else :
            if username == self.__username and password != self.__password:
                print('Bro your password is incorrect !')
                self.borrow()
            else :
                print("Bro invalid username and password ")
                self.borrow()

    def returning(self):
         username = input('Enter your username : ')
         password = int(input('Enter your password : '))
         print(f'Welcome {self.fname} {self.lname} Sir Thanks for Choosing our services !')
         if password == self.__password and username == self.__username and self.book > 0:
             print(f'{self.book} books are deposited successfully ')
             library.__Total_books = library.__Total_books + self.book
             print('Did you want to by any new Book from us ( y/n ) : ')
             response = input('Enter your response : ')
             if response == 'y' or response == 'Y':
                 self.borrow()
             elif response == 'n' or response == 'N':
                 self.menu()
             else:
                 print('Please Enter the (y or n) for your response !')
                 self.menu()
         else :
             if username == self.__username and password != self.__password:
                print('Bro your password is incorrect !')
                self.returning()
             else :
                print("Bro invalid username and password ")
                self.returning()

    def remaining(self):
         username = input('Enter your username : ')
         password = int(input('Enter your password : '))
         print(f'Welcome {self.fname} {self.lname} Sir Thanks for Choosing our services !')
         if password == self.__password and username == self.__username and self.book > 0:
             print(f'Total remaining Books are {library.__Total_books}')

             print('Did you want to explore more in menu (y/n) :')
             response = input('Enter your response : ')
             if response == 'y' or response == 'Y':
                 self.menu()
             else :
                 exit()
         else :
             if username == self.__username and password != self.__password:
                print('Bro your password is incorrect !')
                self.remaining()
             elif self.book == 0:
                 print("You dont Borrowed any Book yet ! ")
                 response = input("Did You want to Borrow Books (y/n) : ")
                 if response == 'y' or  response == 'Y':
                     self.borrow()
                 elif response == 'n' or  response == 'N':
                     self.menu()
                 else :
                     print("Please select the (y/n) only !")
                     self.menu()
             else :
                print("Bro invalid username and password ")
                self.remaining()

    def profile(self):
         username = input('Enter your username : ')
         password = int(input('Enter your password : '))
         if password == self.__password and username == self.__username:
             self.user_profile['Borrowed Books'] = self.book
             print(self.user_profile)
             self.menu()
         else :
             if username == self.__username and password != self.__password:
                print('Bro your password is incorrect !')
                self.profile()
             else :
                print("Bro invalid username and password ")
                self.menu()
```
