```python

class Atm:
    def __init__(self):
        print("Welcome! Please Create Your Account First.")
        self.username = input("Set your New Username: ")
        self.password = input("Set your New Password: ")
        self.balance = 0
        print("Account Created Successfully! \n")
        self.selection()

    def selection(self):

        print('-' * 30)
        choice = input("""
        1. deposit
        2. check
        3. withdraw
        4. exit
        """)
        print('-' * 30)

        if choice == '1':
            self.deposit()
        elif choice == '2':
            self.check()
        elif choice == '3':
            self.withdraw()
        elif choice == '4':
            print('Thanks for Visiting us ! ')
            exit()
        else :
            print('Please Enter the number between 1 to 4 ')
            self.selection()

    def deposit(self):
        username = input("Enter Your username : ")
        password = input('Enter your password  : ')
        if self.username == username and self.password == password:
            amount = float(input('Enter your amount : '))
            self.balance += amount
            print('Your Balance is Succesfully Updated')
            print(f'Our Total balance is {self.balance}')
            self.selection()
        else :
            if username != self.username :
                print('username and password is invaild !')
                try_again = input('Enter y/n : ')
                if try_again == 'y':
                    self.deposit()
                elif try_again == 'n':
                    exit()
                else :
                    print('please enter the y/n words only')
            elif password != self.password:
                print('your password is invaild !')
                self.selection()


    def check(self):
        username = input("Enter Your username : ")
        password = input('Enter your password  : ')
        if self.username == username and self.password == password:
            if self.balance == 0 :
                print("Your bank balance is zero ! (Gareeb) ")
                self.selection()
            else:
                print(f'Total balance is {self.balance}')
                self.selection()
        else :
            if username != self.username :
                print('username and password is invaild !')
                try_again = input('Enter y/n : ')
                if try_again == 'y':
                    self.check()
                elif try_again == 'n':
                    exit()
                else :
                    print('please enter the y/n words only')
            elif password != self.password:
                print('your password is invaild !')
                self.selection()

    def withdraw(self):
        username = input("Enter Your username : ")
        password = input('Enter your password  : ')
        if self.username == username and self.password == password:
            if self.balance == 0 :
                print("Your bank balance is zero ! (Gareeb) ")
                self.selection()
            else:
                withdrawl_amount = float(input("Enter Your Withdrawl Amount : "))
                if withdrawl_amount > self.balance :
                    print('Bro Overdraft facitly is not available')
                    self.withdraw()
                else :
                    self.balance -= withdrawl_amount
                    print('Our Withdraw is Successful !')
                    print(f'Our Withdraw amount is {withdrawl_amount}')
                    self.selection()
        else :
            if username != self.username :
                print('username and password is invaild !')
                try_again = input('Enter y/n : ')
                if try_again == 'y':
                    self.withdraw()
                elif try_again == 'n':
                    exit()
                else :
                    print('please enter the y/n words only')
            elif password != self.password:
                print('your password is invaild !')
                self.selection()


```
