```python
class calculator:
    def __init__(self):
        a = float(input("Enter the first number : "))
        b = float(input("Enter the second number : "))
        op = input("Enter the operator you want ( + , - , %, *, /, **) :  ")
        self.a = a
        self.b = b
        self.op = op
        self.choice()

    def choice(self):

        if self.op == '+' or self.op == '-':
            if self.op == '+':
                print(f'The sum of {self.a} and {self.b} = {self.a + self.b}')
                self.response()
            else:
                print(f'The subtraction of {self.a} and {self.b} = {self.a - self.b}')
                self.response()
        elif self.op == '*' or self.op == '/':
            if self.op == '*':
                print(f'The Multiplication of {self.a} and {self.b} = {self.a * self.b}')
                self.response()
            else:
                print(f'The division of {self.a} and {self.b} = {self.a / self.b}')
                self.response()
        elif self.op == '**' or self.op == '%':
            if self.op == '**':
                print(f'The Square of {self.a} and {self.b} = {self.a ** self.b}')
                self.response()
            else:
                print(f'The mode of {self.a} and {self.b} = {self.a % self.b}')
                self.response()
        else:
            print('invalid choice try again !')
            self.choice()

    def response(self):
        res = input("Are u want use it again ? ( y / n) : ")

        if res == 'y' or res == 'Y' or res == 'yes' or res == 'Yes' or res == 'haan' :
            self.__init__()
        elif res == 'n' or res == 'N' or res == 'no' or res == 'No' or res == 'nahi' :
            exit()
```

```python
if __name__ == '__main__':
    c = calculator()
```
