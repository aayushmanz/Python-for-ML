```python
# Version of calculator using Exception and decorators

def Exception_handler(datatype):
     def wrapper(func):
       def wrapper_for_value(a,b,ops):
         if type(a) != datatype or type(b) != datatype:
            raise TypeError ('Please sir enter the float value only becuz our lazy ayush sir want that !')
         else :
            return func(a,b,ops)
       return  wrapper_for_value
     return wrapper

class calculator:

    def __init__(self, a, b, ops):
        self.a = a
        self.b = b
        self.ops = ops
        self.choice()

    def choice(self):

        if self.ops == '+' or self.ops == '-':
            if self.ops == '+':
             print('{} + {} = {}'.format(self.a, self.b, (self.a + self.b)))

            else :
                print('{} - {} = {}'.format(self.a, self.b, (self.a - self.b)))

        elif self.ops == '*' or self.ops == '/':
            if self.ops == '*':
                print('{} * {} = {}'.format(self.a, self.b, (self.a * self.b)))

            elif self.ops == '/':
               try :
                print('{} / {} = {}'.format(self.a, self.b, (self.a / self.b)))
               except:
                   raise ZeroDivisionError('Bhai zero mat laga yaha pe !')


        elif  self.ops == '%' or self.ops == '//':

            if self.ops == '%':
                print('{} % {} = {}'.format(self.a, self.b, (self.a % self.b)))

            else :
                print('{} // {} = {}'.format(self.a, self.b, (self.a // self.b)))

        else :
            print("Invalid choice ")
            exit()


@Exception_handler(float)
def lets_calculate(x, y, ops):
    solve = calculator(x, y, ops)
    solve.choice()


if __name__ == '__main__':
  while True:
    try:
      x = float(input("Enter the first value : "))
      y = float(input("Enter the second value : "))
      ops = input("Enter the operator : ")
      lets_calculate(x, y, ops)

      response = input("What you want to do ( y / n ) : ")
      if response == 'y' or response == 'Y':
        pass
      elif response == 'n' or response == 'N':
        break
      else :
        print('Bhai dekh ese nahi chalta hai ')
        break
    except KeyboardInterrupt as e:
      print("Bro please decide what you want to so here ! ")
      break
    except ValueError as e :
        print('Bhai pls samjh yaha sirf float kaam karta !')
        break
```
