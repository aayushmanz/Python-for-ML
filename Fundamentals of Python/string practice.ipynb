```python
s = input("Enter the Email : ")
term = input("Enter the term to be searched : ")
counter = 0
for i in s:
    if i == term:
        counter = counter + 1

print("frequency", counter)
```

```python
s = input("Enter the string : ")
term = input("What would you like to remove : ")
result = ''

for i in s:
    if i != term:
        result = result + i
print(result)
```

```python
# palindrome or not
s = input("Enter the string : ")
print(s)
flag = True
for i in range(0,len(s)//2 ):
    if s[i] != s[len(s) - 1 - i]:
        flag = False
        print("palindrome nahi hai")
        break

if flag:
    print("Palindrome")
```
