```python
# Loop control statement :
# 1. Break
# 2. Continue
# 3. Pass
```

```python
# 1. Break
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python
# Example code for Break :

lower = int(input("Enter The lower Number : "))
upper = int(input("Enter the Upper number : "))

for i in range(lower, upper+1):
    for j in range(2, i):
        if i%j == 0:
            break
    else:
         print(i)


```

```python
# 2. Continue

for i in range(1, 10):
    if i == 5:
        continue
    print(i)
```

```python
# 3. pass

for i in range(1, 10):
    pass   # this is kind of helping us for stoping our code without facing any error because of half code.
```
