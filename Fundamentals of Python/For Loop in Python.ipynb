```python
# for Loop Demo :
print("First loop")
for i in range(1, 10):  # range is use for setting the Size of our loop
    print(i)            # Example : range(first value, last value)


print('Second Loop')
for i in range(1, 10, 2): # Example : range(first value, last value, Distance between values)
    print(i)
```

```python
#Program - The current population of a town is 10,000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

cur_pop = 10000
growth = 0.10
for i in range(1, 11):
    cur_pop = cur_pop/(1+growth)
    print(f'Year {i} age : {int(cur_pop)}')

```
