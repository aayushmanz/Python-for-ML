```python
# menu driven calculator
num = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

ops = input("Enter the operator U want : ")

if ops == "+":
    print(num + num2)
elif ops == "-":
    print(num - num2)
elif ops == "*":
    print(num * num2)
elif ops == "/":
    print(num / num2)
elif ops == "%":
    print(num % num2)
else :
    print("Enter the Right ops")
```
