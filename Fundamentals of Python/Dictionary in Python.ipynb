Dictionary

Dictionary in Python is a collection of keys values, used to store data values like a map, which, unlike other data types which hold only a single value as an element.

In some languages it is known as map or associative arrays.

dict = { 'name' : 'ayush' , 'age' : 22 , 'gender' : 'male' }

Characteristics:

Mutable: This means you can change, add, or remove items after the dictionary has been created.

Indexing has no meaning: Unlike lists, dictionaries do not use numerical indices (0, 1, 2) to access elements; instead, they use unique keys.

Keys can't be duplicated: A dictionary cannot have two items with the same key. If you try to assign a new value to an existing key, the old value will be overwritten.

Keys can't be mutable items: Dictionary keys must be of an immutable data type (like strings, numbers, or tuples). You cannot use a list or another dictionary as a key.

Create Dictionary :

```python
# empty dictionary :
d = {}
print(d)

# 1D dict :
d = {
    'name' : 'ayush',
    'class' : 12
     }
print (d)
```

```python
# with mixed dict :
d = {(1,2,3) : 1, 'name' : 'ayush'}
print(d)

# 2D dict :
s={
    'name' : "ayush",
    'college' : 'vision',
    'sem' : 3,
    'subject' : {
        'java' : 79,
        'Os' : 70,
        'AFM' : 67
    }
}
```

```python
# using sequence and dict function :
d = dict([('name','ayush'),('sem', 3),('age' , 22)])
print(d)
# duplicate keys
d = {'name' : 'ayush', 'name' :'mahipal'}
print(d)
# mutable items as keys
d = {'name':'ayush',(1,2,3) : 2}
print(d)
```

Accessing items :

```python
d = {
    'name' : 'ayush',
    'class' : 12
     }
# print(d[0]) # not allowed !
d['class']
d.get('name')
```

```python
# Adding value pairs in dict :
d['gender'] = 'male'
print(d)
```

```python
d = {'name': 'ayush', 'class': 12, 'gender': 'male'}
# pop :
d.pop('class')
print(d)
```

```python
# popitem :
d = {'name': 'ayush', 'class': 12, 'gender': 'male'}
d.popitem()
```

```python
d = {'name': 'ayush', 'class': 12, 'gender': 'male'}
# del
del d['name']
print(d)
```

```python
# clear :
d = {'name': 'ayush', 'class': 12, 'gender': 'male'}
d.clear()
print(d)
```

```python
# accessing from 2D dictionary

s={
    'name' : "ayush",
    'college' : 'vision',
    'sem' : 3,
    'subject' : {
        'java' : 79,
        'Os' : 70,
        'AFM' : 67
    }
}
s['subject']['Os']
```

```python
# Adding/editing key value pair :
s['subject']['English'] = 77
print(s)

# del s['subject']['java']
```

Dict Operations :

1. membership
2. iterations

```python
s={
    'name' : "ayush",
    'college' : 'vision',
    'sem' : 3,
    'subject' : {
        'java' : 79,
        'Os' : 70,
        'AFM' : 67
    }
}

print(s)

# 'ayush' in s    # its always works on keys not in values
'name' in s
```

```python
d = {'name': 'ayush', 'class': 12, 'gender': 'male'}

for i in d:
    print(i," : ",d[i])
```

Dictionary / Function :

```python
# len / sorted :
d = {'name': 'ayush', 'class': 12, 'gender': 'male'}
len(d)
print(d)
```

```python
d = {'name': 'ayush', 'class': 12, 'gender': 'male'}
sorted(d, reverse=True)
```

```python
d = {'name': 'ayush', 'class': 12, 'gender': 'male'}
max(d)
```
