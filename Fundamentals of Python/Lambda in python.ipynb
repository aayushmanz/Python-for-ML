Lambda Function:

A lambda function is a small anonymous function.
A lambda Function can take any number of arguemnts, but can only have one expression.

1. Example :
     lambda a,b : a+b
2. Syntax :
     lambda parameter : expression

```python
# x -> x^2
a=lambda x:x**2
print(a(2))
```

```python
# x,y -> x+y
a = lambda x,y : x*y
print(a(2,3))
```

Diff between lambda vs Normal Function :

 1. No name
 2. lambda has no return value(infact,return a function)
 3. lambda is written in 1 line
 4. not reusable
1. Then why use lambda functions.
2. They are used in  HOF (higher order functions)

```python
# check if a string has 'a'

a = lambda s : 'a' in s
a('ayush')
```

```python
# odd or even :
a = lambda x:'even' if x%2 == 0 else 'odd'
print(a(6))
```

Higher order Function :

```python
# example :

def sq(x):
    return x**2
def cube(x):
    return x**3
def transform(f,l):
    output = []
    for i in l:
        output.append(f(i))

    print(output)

l = [1,2,3,4,5]
transform(lambda  x:x**3,l) # so the ans of X = 1 : X = l = 1 ** 3 ( 1 * 1 * 1 )
```

Map :

```python
# square the items of a list
list(map(lambda x : x**2,[1,2,3,4,5]))
```

```python
# odd / even labelling a list items
list(map(lambda x : 'even' if x % 2 == 0 else 'odd',[1,2,3,4,5]))
```

```python
# fetch names from a list of dict :

users = [
    {
        'name':'ayush',
        'age':22,
        'gender':'male'
    },
    {
        'name':'ladki',
        'age':30,
        'gender':'female'
    },
    {
        'name':'lucky',
        'age':78,
        'gender':'trans'
    }
]

result = list(map(lambda user : user['name'],users))
print(result)
```

Filter :

```python
# numbers greater than 5
l = [3,4,2,7,8,9]

list(filter(lambda x : x > 5, l))
```

```python
# fetch fruits starts with ' a '

fruits = ['apple', 'guava','mango','cherry']

list(filter(lambda x : x.startswith('a'), fruits))
```

Reduce :

```python
# sum of all items :
import functools

functools.reduce(lambda x,y : x+y,[1,2,3,4,5])
```

```python
# find min :
functools.reduce(lambda x,y : x if x < y else y,[12,45,67,16,99,46] )
```

```python
# find max :
functools.reduce(lambda x,y : x if x > y else y,[12,45,67,16,99,46] )
```
