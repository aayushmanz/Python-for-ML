3. Changeble / mutable

```python
# create a list :
# 1. empty
print([])
# 1. 1D list - > homogeneous
print([1,2,3,4,5])
# 3. 2D list -> nested list
print([1,2,3,[4,5]])
# 4. 3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
# 5. Heterogeneous
print([1, True, 5.6,5+6j,'ayush'])
# 6. using type Conversion
print(list('ayush'))
```

Accessing Items from a list.

```python
# 1. indexing
l = [1,2,3,4,5]
print(l[0]) # -> positive indexing
print(l[-1])# -> negative indexing

# -> 2D indexing Example
l = [1,2,3,[4,5]]
print(l[-1][-2])
print(l[3],[4])

# -> 3D indexing Example
l = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(l[1][0][1])
print(l[0][0][1])
```

```python
# Slicing
l = [1,2,3,4,5]
print(l[0:2])
print(l[0::3])# For jump we use (::)
print(l[::-1])# index become negative (Reverse)
```

Adding Items to a list :

```python
# 1. append
l = [1,2,3,4,5,6]
l.append(True)
l.append([7,8,9]) # If we use append like that then we are creating heterogeneous list
print(l) # append always try to add single value.
```

```python
# 2. extend
l = [1,2,3,4,5,6]
l.extend([7,8,9])
l.extend("ayush") # Extend always try to split the value in pieces
print(l)
```

```python
# 3. insert
l = [1,2,3,4,5,6]
l.insert(2,67)
l.insert(6,"ayush")
print(l)
```

Editing items in a list

```python
l = [1,2,3,4,5,6]

l[-1] = 500
l[0] = 100
l[1:4]=["ayush","mahipal","deepak"]
print(l)
```

Deleting items from the list.

```python
# 1. del -> list is mutable
l = [1,2,3,4,5,6]
print(l)
del l
# print(l) -> it will show the error Because the list is deleted.
```

```python
l = [1,2,3,4,5,6]
del l[-1]
print(l)
```

```python
l = [1,2,3,4,5,6]
del l[1:3]
print(l)
```

```python
# 2. remove
l = [1,2,3,4,5,6]
l.remove(3)
print(l) # now Three is deleted !
```

```python
# 3. pop

l = [1,2,3,4,5,6]
l.insert(5, "ayush")
l.insert(7,"Vatsal")
print(l)
l.pop(5) # 5 means 5th index !
print(l)
```

```python
# 4. clear
l = [1,2,3,4,5,6]
l.clear() # clear use to remove the data from the list (l) but it not delete the whole list.
print(l) # it will show empty list
```

Operations on lists :
1. Arithmetic
2. Membership
3. Loop

```python
# 1. Arithmetic
l1 = [1,2,3,4]
l2 = [5,6,7,8]

print(l1 + l2)
```

```python
print(l1*3)
```

```python
# 2. Membership
l1 = [1,2,3,4,5]
l2 = [1,2,3,4,[5,6]]

print(5 not in l1)
print(5 in l2)
print([5,6] in l2)
```

```python
# 3. Loops
l1 = [1,2,3,4,5]
l2 = [1,2,3,4,[5,6]]

for i in l1:
    print(i)
print()
for i in l2:
    print(i)
```

```python
l3 = [[[1,2],[3,4]],[[5,6],[7,8]]]
for i in l3:
    print(i)
```

List Functions :
1. count -> len(), min(), max()
2. index
3. reverse
4. sort ( vs sorted )
5. copy

```python
l = [5,6,3,8,2,9,7]
print(len(l))
print(min(l))
print(max(l))
print(sorted(l))
print(l)   # -> temp
```

```python
# count :
l = [1,2,3,4,5,5,5,5]
l.count(5)
```

```python
# index :
l = [1,2,3,4,5]
l.index(4)
```

```python
# reverse :
l = [1,2,3,4,5]
l.reverse()
print(l)
```

```python
# sort ( vs Sorted ) :
l = [5,6,3,8,2,9,7]
print(l)
print(sorted(l)) # sorted sort for temp time.
print(l)
l.sort() # sort is permanent
print(l)
print(l)
```

```python
# copy -> shallow
l = [5,6,3,8,2,9,7]
print(l)
print(id(l))
l1 =l.copy()
print(l1)
print(id(l1))
```
