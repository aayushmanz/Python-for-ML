```python
# First way to do sequence sum.

n = int(input("Enter the N value : "))

sum = 0
for i in range(1, n+1):
     sum += i
     print(sum)
```

```python
# 1/1! + 2/2! + #3/3! ....

n = int(input("Enter the N value : "))
fact = 1
value = 0.0
for i in range(1, n+1):
    fact = fact * i
    value = value + i / fact
    print(value)
```
