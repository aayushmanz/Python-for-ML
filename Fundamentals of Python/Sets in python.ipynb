Set :

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).
However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characteristics:

1. Unordered
2. Mutable
3. No Duplicates
4. Can't contain mutable data types

Creating sets :

```python
# empty set :
s = set()
print(s)
print(type(s))
```

```python
# 1d set :
s = {1,2,3}
print(s)
print(type(s))
```

```python
# 2d sets :
s = {1,2,3,{4,5}}
print(s) # not allowed !
```

```python
# homo and hetro
s = {1,'hello',2.3,True,(7,8,9)}
print(s)
```

```python
# using Type conversion :
s = set([1, 2, 3, 4])
print(s)
```

```python
# duplicates not allowed :
s = {1,2,3,3,4,4,5}
print(s)
```

```python
# set can't have multiple items :
s = {1,2,3,[4,5]}
print(s)
```

```python
# order never matters in sets :

s = {1,2,3}
s1 = {3,2,1}

print(s == s1)
```

Accessing items :

```python
s = {1,2,3,4}
s1[0:2] # not allowed accessing items.
```

Editing items :

```python
s = {1,2,3,4}
s[0] = 500
# sets are immutable
```

Adding items :

```python
s =  {1,2,3,4}

# add :
s.add(5)
print(s)

# update :
s.update([5,6,7,8])
print(s)

# del :
s =  {1,2,3,4}
print(s)
del s
# print(s) -. deleted !

# discard :
s =  {1,2,3,4,5}
print(s)
s.discard(5)
print(s)

# remove :
s.remove(3)
print(s)

# pop : -> randomly delete number :
s.pop()
print(s)

# clear :
s.clear()
print(s)
```

Sets  Operations :

1. union(|)
2. Intersection(&)
3. Difference(-)
4. Symmetric Difference(^)
5. Membership test
6. Iteration

```python
# Union :
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1 | s2
```

```python
# intersection :
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1 & s2
```

```python
# difference (-)
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1 - s2
```

```python
# Symmetric difference (^)
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1 ^ s2
```

```python
# Membership :
1 in s1
1 not in s1
```

```python
# iterations :
for i in s1 :
    print(i)
```

Len/ sum/ min/ max/ sorted :

```python
# len() :
s = {3,1,2,5,8,9}
print(len(s))
```

```python
# sum :
print(sum(s))
```

```python
# min :
print(min(s))
```

```python
# max :
print(max(s))
```

```python
# sorted :
print(sorted(s))
print(sorted(s, reverse=True))
```

```python
# union / update :
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 | s2
s1.union(s2)

s1.update(s2)
print(s1)
print(s2)
```

```python
# intersection / Intersection_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.intersection(s2)
s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference / difference_update :
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.difference(s2)
s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference / symmetric_difference_update :
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.symmetric_difference(s2)
s1.symmetric_difference_update(s2)

print(s1)
print(s2)
```

```python
# isdisjoint/issubset/issuperset
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.isdisjoint(s2)

```

```python
# issubset :
s1 = {1,2,3,4,5}
s2 = {4,5}

s1.issubset(s2)
```

```python
# issuperset :
s1 = {1,2,3,4,5}
s2 = {4,5}

s1.issuperset(s2)
```

```python
# copy :

s1 = {1,2,3,4,5}
s2 = s1.copy()
print(s1)
print(id(s1))
print(s2)
print(id(s2))
```
