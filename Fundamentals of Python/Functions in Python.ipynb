Lets Create a function. ( with docstring )

```python
def is_even(num):
    """
    Docstring content...
    """
    if type(num) == int:
        if num % 2 == 0:
            return 'even'
        else:
            return 'odd'
    else:
        print("Enter the integer value")
```

```python
# function
# function_name(input)
for i in range(1,11):
    x = is_even(i)
    print(x)
```

*args and **kwargs

```python
# args -> arguments
# allows us to pass a variable number of non - keyword argument to a function.

def multi(*args):
    a = 1
    """
    this is really useful for adding multiple arguments in the parenthesis.
     """
    for i in args:
        a = a * i
    print(args)
    return a

```

```python
multi(2,3,5,6,7)
```

```python
multi.__doc__
```

```python
# **kwargs
# **kwargs allows us to pass number of keyword arguments.
# keyword arguments mean that they contain a key-value pair, like a python dictionary.
def display(**kwargs):
    for(key,value) in kwargs.items():
        print(key,'->',value)
```

```python
display(india='delhi',nepal='kathmandu',pakistan='Terrorists')
```

Points to remember while using *args and **kwargs

1. order of the arguments matter(normal -> *args -> **kwargs)
2. the words 'args' and 'kwargs' are only convertion, you can use any name of your choice !

Variable scope :

```python
# example : Local variable can take global variable value but global variable cannot be able to take value of local.
def f(y):
    x = 1 # this is local variable
    x += 1
    print(x)

x = 5 # this is global variable.
f(x)
print(x)
```

```python
# example :
def h(y):
    x += 1 # can not change the global variable through local variable.

x = 5
h(x)
print(x)


```

```python
# but we can do this too :
def h(y):
    global x  # because of this we can forcefully change it.
    x += 1

x = 5
h(x)
print(x)
```

```python
# example :

def f(x):
    x = x+1
    print('in f(x): x =', x)
    return x

x = 3
z = f(x)
print('in main program scope : z =', z)
print('in main program scope : x =', x)
```

Nested Functions :

```python
# example :
def f():
    def g():
        print('inside function g')
    g()
    print('inside function f')
```

```python
f() # we not call the inner function g() outside the main function.
```

```python
# example :
def g(x):
    def h():
        x = 'abc'
    x = x + 1
    print('in g(x) : x =', x)
    h()
    return x

x = 3
z = g(x)
```

Functions are 1st class citizen :

```python
# type and id
def square(num):
    return num**2
type(square) # function
id(square)
```

```python
# reassign
x = square
id(x) # id will show
x(3)
```

```python
# deleting a function
del square # we can also delete the function like datatype.
```

```python
# storing :
def square(num):
    return num**2
type(square) # function
id(square)

l = [1,2,3,4,square]
l[-1](3) # square(3)
```

```python
# returning a function.
s = {square} # immutable becz set will not allow mutable datatypes.
s
```

```python
# example :

def f(): # (1)
    def x(a,b): # (3)
        return a+b
    return x # we are returning a function now # (4)
val = f()(3,4) # f() will replace by the function x -> x(3,4) # (2) # (5) but now it is converted into x
print(val) # (6)
```

```python
# example :

def fun_a():
    print('inside fun_a')

def fun_b(z):
    print('inside fun_b')
    return z()
print(fun_b(fun_a))
```
