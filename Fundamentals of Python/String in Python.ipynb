```python
# Create Strings
s ="HELLO"
s ='HELLO'
# Multiline Strings
s ='''HELLO'''
s ="""HELLO"""
s = str('HELLO')
print(s)
```

```python
#indexing
s = "hello ayush !"
print(s[7])
```

```python
# negative indexing
s = "hello ayush !"
print(s[-7])
```

```python
# example for negative indexing

s = "Hello mahipal"
print(s[-8:])
print(s[-11:-2])
print(s[-11:-1:2])
```

```python
# Slicing
s = "Hello ayush Bhai"
print(s[6:9])
print(s[6:])
print(s[0:12:2]) # 2 -> size of jump between words
print(s[8:0:-2]) # now the sentence will pick by the index will be opposite
print(s[::-2])
```

```python
# editing & deleting String
# python strings are immutable ( can not change the String in python )

s = "Hello ayush Bhai"
del s
# print(s)
# this will not work because String is deleted !
```

```python
# Example for what is not allowed in negative indexing
s = "hello ayush !"
# del s[-1:-5:2] -> this is not allowed !
print(s)
```
