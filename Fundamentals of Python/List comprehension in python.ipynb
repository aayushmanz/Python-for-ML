list comprehension :

list comprehension provide way of creating lists.

formula :
newlist = [expression for item in iterable if condition == True]


Advantanges of list comprehension :

1. more time-efficient and space-efficient than loops.
2. require fewer lines of codes.
3. transforms iterative statements into a formula.

```python
from decorator import append

# add 1 to 10 numbers to a list :
# normal way :
l = []
for i in range(1, 11):
    l.append(i)
print(l)
```

```python
# new method of list comprehension
l = [i for i in range(1, 11)]
print(l)
```

```python
# scalar multiplication on a vector
# normal way

v = [2,3,4]
s = -3

# [-6, -9, -12]
[s * i for i in v]


```

```python
# add squares :
l = [1,2,3,4,5]
[i**2 for i in l]   # [i*i for i in l]
```

```python
# print all number which are divisible by five in the range of 1 to 50
[i for i in range(1, 51) if i%5 == 0]
```

```python
# find language which start with letter p
languages = ['java','python','php','c','javascript']
[language for language in languages if language.startswith('p')]
```

```python
#  nested if with list comprehension :
basket = ['apple', 'banana', 'kiwi', 'cherry']
my = ['apple', 'gulab jamun', 'rasmalai', 'barfi']

[i for i in my if i in basket if i.startswith('a')]
```

```python
# print a (3,3) matrix using list comprehension -> nested list comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# Cartesian products -> list comprehension on 2 lists together
l1 = [1,2,3,4]
l2 = [5,6,7,8]

[i*j for i in l1 for j in l2]
```

2 ways to traverse a list

1. itemwise
2. indexwise

```python
# itemwise
l = [1,2,3,4]

for i in l:
    print(i)
```

```python
# indexwise
l = [1,2,3,4]

for i in range(0, len(l)):
    print(i)
```

Zip:

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator in paired together, and then the second item in each passed iteartor are paired together.

```python
l1 = [1,2,3,4]
l2 = [-1,-2,-3,-4]

list(zip(l1,l2))
[i+j for i,j in zip(l1,l2)]
```
