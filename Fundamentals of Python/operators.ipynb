```python
# arithmatic operators
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print("The values are ",a," or ",b)
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
print(a**b)
```

```python
# relational operators
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print("The values are ",a," or ",b)
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
#logical operators
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print("The values are ",a," or ",b)
print(a > 0 and b > 0)
print(a > 0 or b > 0)
print(not (a > 0 and b > 0))
```

```python
#bitwise operators
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print("The values are ",a," or ",b)
print(a & b)  # bitwise and operator
print(a | b)
print(a ^ b)
print(~a)
print(a << 1)
print(a >> 1)
```

```python
#assignment operators
a = int(input("Enter the first number: "))
print("The value is ",a)
a += 5
print(a)
a -= 5
print(a)
a *= 5
print(a)
a /= 5
print(a)
a //= 5
print(a)
a %= 5
print(a)
a **= 5
print(a)

```

```python
#membership operators
print("d" in "delhi")
print("d" not in "delhi")
print(1 in [1,2,3,4,5])
print(1 not in [1,2,3,4,5])
```
