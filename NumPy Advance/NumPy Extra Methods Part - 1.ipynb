# np.sort

Return a sorted copy of an array.

```python
# code :

import numpy as np

a = np.random.randint(1,100,15)
print(a)
```

```python
b = np.random.randint(1,100, 24).reshape(6,4)
print(b)
```

```python
np.sort(a) # ascending order
```

```python
np.sort(b) # row wise sorting # 2D
```

```python
np.sort(b, axis=0)
```

```python
np.sort(a)[::-1] # we can also use indexing in sorting !
```

# np.concatinate :

* numpy.concatenate() function concatenate a sequence of arrays along an existing axis.

```python
# code :

c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)
```

```python
np.concatenate((c,d), axis=1) # column
```

```python
np.concatenate((c,d), axis=0) # row
```

# np.where() :

* The numpy.where() function returns the indices of elements in an input array where the given condition is satisfied.

```python
# code :
a
```

```python
# find all indices with value greater than 50

np.where(a > 50)
```

```python
# replace all values > 50 with 0
np.where(a>50,0, a) # replacing the whole greater 50 value with 0
```

```python
np.where(a%2 == 0,0, a) # finding all even indexes
```

# np.cumsum() :

* The numpy.cumsum() function is used when we want to computer the cumulative sum of array elements over a given axis.

```python
a
```

```python
np.cumsum(a) # this is use for calculating all elements in array index ( + )
```

```python
b
```

```python
np.cumsum(b) # without axis it shows the output in 1D array
```

```python
np.cumsum(b, axis=1) # row sum
```

```python
np.cumsum(b, axis=0) # column sum
```
