# Set Functions :
* np.union1d
* np.intersect1d
* np.setdiff1d
* np.setxor1d
* np.in1d

```python
import numpy as np

m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])
np.union1d(m,n) # jointing and adjusting the duplicate values
```

```python
np.intersect1d(m,n) # Similar values in both arrays
```

```python
np.setdiff1d(m,n) # those items which are not in 'n'.
```

```python
np.setdiff1d(n,m) # showing the values of m now which are not present.
```

```python
np.setxor1d(m,n) # Those elements which are not similar in both arrays.
```
