# Plotting Graphs :

```python
# plotting a 2D plot
# x = y

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)
y = x

plt.plot(x,y)
```

```python
# y = x^2

x = np.linspace(-10,10,100)**2
y = x
plt.plot(x,y)
```

```python
# y = sin(x)
x = np.linspace(-10,10,100)
y = np.sin(x)

plt.plot(x,y)
```

```python
# y = xlog(x)
x = np.linspace(-10,10,100)
y = x * np.log(x)

plt.plot(x,y)
```

```python
# sigmoid
x = np.linspace(-10,10,100)
y = 1/(1+np.exp(-x))

plt.plot(x,y)
```
