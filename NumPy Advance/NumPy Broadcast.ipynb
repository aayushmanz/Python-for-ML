# Broadcasting  :

1. **What is Broadcasting in NumPy?**

* Broadcasting is a mechanism in NumPy that allows arrays of different shapes to be used together in arithmetic operations.

* It automatically expands the smaller array to match the shape of the larger array without actually copying data.

```python
# same shape
import numpy as np

a = np.arange(6).reshape(2,3)
b = np.arange(6).reshape(2,3)

print(a)
print(b)

print(a + b)
```

```python
# diff shape :

a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print(a + b)
```

```python
# random example :
a = np.random.randint(1,100,6).reshape(2,3)
b = np.random.randint(1,100,6).reshape(2,3)

print(a)
print(b)

print(a + b)
```

# Broadcasting Rules :

* _Broadcasting in NumPy is a mechanism that allows operations on arrays of different shapes. It works by automatically expanding the smaller array to match the shape of the larger array, enabling element-wise computations without copying data._

**Rule 1:** _Compare dimensions from right to left_

NumPy compares the shapes of arrays starting from the last dimension (rightmost side).

**Rule 2:** _Dimensions are compatible if:_

Two dimensions are compatible when:

* They are equal
OR
* One of them is 1

**Rule 3:** _If dimensions are not compatible → Error_

If neither condition is satisfied, NumPy will raise a ValueError.

**Rule 4:** _Missing dimensions are treated as 1_

If arrays have different numbers of dimensions, the smaller one is padded with 1s on the left side.

```python
# correct way for using !

import numpy as np

a = np.arange(1).reshape(1)
b = np.arange(4).reshape(2,2)   # it is valid
print(a)
print(b)
print(a + b)
```

```python
# this is not a correct way

import numpy as np

a = np.arange(1,4).reshape(3)
b = np.arange(12).reshape(3,4) # it is not valid because of last number 4 is greater than 3
print(a)
print(b)
print(a + b)
```

```python
a = np.arange(12).reshape(4,3)
b = np.arange(12).reshape(3,4)   # it is not valid
print(a)
print(b)
print(a + b)
```

```python
a = np.arange(16).reshape(4,4)
b = np.arange(4).reshape(2,2)   # it not is valid
print(a)
print(b)
print(a + b)
```
