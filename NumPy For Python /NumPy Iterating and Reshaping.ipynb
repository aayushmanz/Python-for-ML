**Iterating :**

```python
# Old way :

import numpy as np

a1 = np.arange(10)
for i in a1:
    print(i)
```

```python
# New way by using np.nditer()

for i in np.nditer(a1):
    print(i)
```

```python
# Old way :
a2 = np.arange(12).reshape(3,4)
for i in a2:
    for j in i:
      print(j)
```

```python
# New way by using np.nditer()

for i in np.nditer(a2):
    print(i)
```

```python
# Old way :

a3 = np.arange(27).reshape(3,3,3)

for i in a3:
    for j in i:
        for k in j:
            print(k)
```

```python
# New way by using np.nditer()

for i in np.nditer(a3):
    print(i)
```
