**Stacking :**

```python
# horizontal Stacking
import numpy as np

a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
```

```python
np.hstack((a4,a5))
```

```python
# Vertical Stacking

np.vstack((a4,a5))
```

**Splitting :**

```python
# horizontal Splitting

np.hsplit(a4,2)
```

```python
# Vertical Splitting
a6 = np.vstack((a4,a5))
np.vsplit(a6,2)
```
