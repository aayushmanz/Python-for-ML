**Abstraction :**

```python
from abc import ABC, abstractmethod
class BankApp(ABC):

    def database(self):
        print('connected to databases !')

    @abstractmethod
    def security(self):
        pass
```

```python
class mobileapp(BankApp):

    def mobile_login(self):
        print('login into mobile')

    def security(self):
        print('mobile Security ! ')
```

```python
mob = mobileapp()
```

```python
mob.security()
```

```python
mob.database()
```
