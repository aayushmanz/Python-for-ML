**Encapsulation :**

**Encapsulation** _is one of the core principles of Object-Oriented Programming (OOP). It refers to the concept of wrapping data (variables) and methods (functions) together into a single unit, usually a class, and restricting direct access to some of the object's components. This helps protect data from accidental modification and improves code security and maintainability._

```python
class Atm:  # class is a blueprint.
    # constructor
    def __init__(self):
        self.pin = ''
        self.__balance = 0  # __ is use for private our variable
        print("I am the constructor!")
        self.menu()
    def get_balance(self): # getter and setter gives use the way to use the private variable access.
        return self.__balance

    def set_balance(self, new_value):
        if type(self.__balance) == int:
           self.__balance = new_value
        else :
            print('bhai integer value hi dal skte hai !')
    def menu(self):
        user_input = input("""
        Hi, how can I help you?
        1. Create pin
        2. Change pin
        3. Check balance
        4. Withdraw
        5. Exit
        """)

        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw()
        else:
            exit()

    def create_pin(self):
        user_pin = input('Enter your pin: ')
        self.pin = user_pin

        user_balance = int(input("Enter your balance: "))
        self.__balance = user_balance

        print("Pin is created successfully!")
        self.menu()

    def change_pin(self):
        old_pin = input("Enter old pin: ")

        if old_pin == self.pin:
            new_pin = input('Enter new pin: ')
            self.pin = new_pin
            print('Pin change successful!')
            self.menu()
        else:
            print('Nhi kar sakte aap change!')
            self.menu()
    def check_balance(self):
        user_pin = input("Enter your pin : ")
        if user_pin == self.pin:
            print(f"Your account balance is : {self.__balance} ")
        else:
            print('Please enter the right pin !')

    def withdraw(self):
        user_pin = input('Enter the pin : ')
        if user_pin == self.pin:
           amount = int(input("Enter the withdraw amount : "))
           if amount <= self.__balance:
               self.__balance = self.__balance - amount
               print('withdrawl succesful, balance is ', self.__balance)
           else :
               print("Not sufficient balance ! ")
        else:
            print('Your pass is incorrect !')
        self.menu()
```

```python
# list of object :

class person :
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender

p1 = person('ayush', 'male')
p2 = person('vatsal', 'female')

l = [p1, p2]

for i in l:
    print(i.name, i.gender)
```

```python
# dict of object :

class person :
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender
    def __str__(self):
        '{} / {}'.format(self.name, self.gender)

p1 = person('ayush', 'male')
p2 = person('vatsal', 'female')

d = {'p1': p1, 'p2': p2}

for i in d:
    print(d[i].gender)
```
