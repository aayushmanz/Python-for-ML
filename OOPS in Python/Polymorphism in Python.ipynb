```python
# Example of Method Overloading :
 # In Method overloading our user decide the method on the basis of parameters.
class shape:

    def area(self, radius):
        return 3.14*radius**2

    def area(self, l, b):
        return l*b

s = shape()
# s.area(45) # taking radius now because of one parameter
# but in python there is no way to use it .
# because python never allows Method overloading just like java don't allow multiple inheritance !
s.area(23, 45) # l * b because our parameters are two
```

```python
# but there is one hack to use the method overloading in python

class shape:

    def area(self, a, b = 0):
        if b == 0:   # because of condition we can kind of use method overloading !
            return 3.14*a**2
        else :
            return a + b

s = shape()
print(s.area(2))
print(s.area(3,4))
```

```python
# Examples for method overriding :

class Animal:
    def speak(self):
        print("The animal makes a sound")

class Dog(Animal):
    # Overriding the speak method
    def speak(self):
        print("The dog barks: Bhaauuu! Bhauuuu!")

class Cat(Animal):
    # Overriding the speak method
    def speak(self):
        print("The cat meows: Meow!")

# Usage
my_dog = Dog()
my_dog.speak()  # Output: The dog barks: bhaauuu! bhaauuuu!
```

```python
# more Example :
class Payment:
    def process_payment(self, amount):
        print(f"Processing a generic payment of ${amount}")

class CreditCardPayment(Payment):
    def process_payment(self, amount):
        # Specific logic for CC (connecting to bank API)
        print(f"Charging ${amount} to Credit Card. Applying 2% transaction fee.")

class CryptoPayment(Payment):
    def process_payment(self, amount):
        # Specific logic for Blockchain
        print(f"Transferring {amount} units via Blockchain wallet.")

# Both objects use the same method name, but behave differently
payments = [CreditCardPayment(), CryptoPayment()]
for p in payments:
    p.process_payment(100)
```
