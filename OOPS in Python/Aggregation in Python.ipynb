**Aggregation (has-a relationship) :**

**Aggregation (OOP Concept)**

_Aggregation is a special form of association in Object-Oriented Programming where one class contains a reference to another class, but both classes can exist independently. It represents a "has-a" relationship._

**In simple words:**
_If object A has object B, and B can still exist without A, that is aggregation._

```python


class customer :
    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender
        self.address = address
    def print_address(self):
        print(self.address.get_city(), self.address.pin, self.address.state)

    def edit_profile(self, new_name, new_city, new_pin, new_state):
        self.name = new_name
        self.address.edit_address(new_city, new_pin, new_state)


class address:
    def __init__(self, city, pin, state ):
        self.__city = city
        self.pin = pin
        self.state = state

    def get_city(self):
        return self.__city
    def edit_address(self, new_city, new_pin , new_state):
        self.__city = new_city
        self.state = new_state
        self.pin = new_pin

add = address('chittorgarh',312001,'rajasthan')
cust = customer('ayush','male',add)  # this is aggregation
cust.print_address()
cust.edit_profile('raju','mumbai',111111,'maharastra')
cust.print_address()
```
