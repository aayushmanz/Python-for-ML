**1. The Class: The Blueprint**
_A Class is essentially a blueprint, template, or a set of instructions for creating something. It doesn't actually exist as a physical thing in your program; it just defines what something should look like and what it should do._

**Real-world analogy:** _Think of a class like a blueprint for a house. You can't live in a blueprint, but it tells the builders exactly how to construct the house, where the doors go, and how many rooms it has._

**2. The Object: The Real Thing**
_An Object is the actual, usable thing created from the blueprint (the Class). The process of creating an object from a class is called instantiation (because the object is an "instance" of the class)._

**Real-world analogy:** _If the class is the blueprint, the object is the actual physical house you build from it. You can build many houses (objects) from the same blueprint (class). They might share the same layout, but one could be painted blue and another painted red._

```python
class Atm:  # class is a blueprint.
    # constructor
    def __init__(self):
        self.pin = ''
        self.balance = 0
        print("I am the constructor!")
        self.menu()

    def menu(self):
        user_input = input("""
        Hi, how can I help you?
        1. Create pin
        2. Change pin
        3. Check balance
        4. Withdraw
        5. Exit
        """)

        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw()
        else:
            exit()

    def create_pin(self):
        user_pin = input('Enter your pin: ')
        self.pin = user_pin

        user_balance = int(input("Enter your balance: "))
        self.balance = user_balance

        print("Pin is created successfully!")
        self.menu()

    def change_pin(self):
        old_pin = input("Enter old pin: ")

        if old_pin == self.pin:
            new_pin = input('Enter new pin: ')
            self.pin = new_pin
            print('Pin change successful!')
            self.menu()
        else:
            print('Nhi kar sakte aap change!')
            self.menu()
    def check_balance(self):
        user_pin = input("Enter your pin : ")
        if user_pin == self.pin:
            print(f"Your account balance is : {self.balance} ")
        else:
            print('Please enter the right pin !')

    def withdraw(self):
        user_pin = input('Enter the pin : ')
        if user_pin == self.pin:
           amount = int(input("Enter the withdraw amount : "))
           if amount <= self.balance:
               self.balance = self.balance - amount
               print('withdrawl succesful, balance is ', self.balance)
           else :
               print("Not sufficient balance ! ")
        else:
            print('Your pass is incorrect !')
        self.menu()
```
