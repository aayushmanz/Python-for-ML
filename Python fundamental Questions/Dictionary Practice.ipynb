```python
dataset = [150, None, 200, 180, None, None, 210]
clean_dataset = {'Valid' : 0, 'Missing' : 0}

for i in dataset:
    if i == None :
        clean_dataset['Missing'] = clean_dataset['Missing'] + 1
    else :
        clean_dataset['Valid'] = clean_dataset['Valid'] + 1

print(clean_dataset)
```

```python

ml_data = {
    "age": 25,
    "income": 0,
    "name": "Ayush",
    "score": 50
}

def process_data(data_dict, key):
    try:
       ans = 100 / data_dict[key]
       print('Answer is : ',ans)
    except KeyError:
        print("KeyError !!")
    except ValueError:
        print('Value Error !!')
    except ZeroDivisionError:
        print('Zero is not divide with any value ! ')
    except TypeError:
        print("Type Error !!")
process_data(ml_data, "name")
process_data(ml_data, "income")
process_data(ml_data, "age")
process_data(ml_data, "score")


```
