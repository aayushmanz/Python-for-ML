```python
# The Data Counter (Dictionary)

votes = ["BJP", "INC", "BJP", "AAP", "INC", "BJP"]
total = {}
for i in votes :
    if i in total:
         total[i] = total[i] + 1
    else :
        total[i] = 1

print(total)



```

```python
dataset = [150, None, 200, 180, None, None, 210]
clean_dataset = {'Valid' : 0, 'Missing' : 0}

for i in dataset:
    if i == None :
        clean_dataset['Missing'] = clean_dataset['Missing'] + 1
    else :
        clean_dataset['Valid'] = clean_dataset['Valid'] + 1

print(clean_dataset)
```
