```python
from parso.python.tree import Number


# The ML Blueprint (OOPs)

class SimpleModel:
    def __init__(self, data):
        print('Ayush ka AI')
        self.data = data

    def predict(self):
        return "{} ** {} = {}".format(self.data, 2, (self.data ** 2)) # square hi toh nikal na tha


if __name__ == '__main__':
    value = int(input("Enter the value : "))
    AI_model = SimpleModel(value)
    print(AI_model.predict())

```

```python
class DataCleaner:
    def __init__(self, raw_data):
        self.raw_data = raw_data

    def get_clean_data(self):
        clear_data = [i for i in self.raw_data if type(i) == int]
        return clear_data




if __name__ == '__main__' :
    dirty_list = [10, "hello", 20, "maverick", 30, "bug", 40]
    value  = DataCleaner(dirty_list)
    print(value.get_clean_data())
```
