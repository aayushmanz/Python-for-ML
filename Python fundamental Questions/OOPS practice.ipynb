```python
# The ML Blueprint (OOPs)

class SimpleModel:
    def __init__(self, data):
        print('Ayush ka AI')
        self.data = data

    def predict(self):
        return "{} ** {} = {}".format(self.data, 2, (self.data ** 2)) # square hi toh nikal na tha


if __name__ == '__main__':
    value = int(input("Enter the value : "))
    AI_model = SimpleModel(value)
    print(AI_model.predict())

```
