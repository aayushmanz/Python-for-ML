```python
# The Accuracy Checker (Exception Handling)

def ayushKaDecorator(funct):
    def wrapper(c, t):
        try :
            return funct(c, t)
        except ZeroDivisionError as e :
            print('Bro Zero se agar divide hota toh tu aaj developer hota !')
    return wrapper

@ayushKaDecorator
def solution(correct, total):
    return (correct / total) * 100


if __name__ == '__main__':
   try :
    correct = float(input("Enter the Correct : "))
    total = float(input("Enter the Total : "))
    res = solution(correct, total)

    print(res)
   except InterruptedError as e :
       print('Han Bhai khel ne apne mein phele !!!!')
```
