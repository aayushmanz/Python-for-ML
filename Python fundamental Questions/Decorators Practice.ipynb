```python

def auto_float(func):
    def wrapper(*args):
        clean_args = []
        for i in args:
          clean_args.append(float(i))
        return func(*clean_args)
    return wrapper


@auto_float
def add_numbers(a, b):
    return a + b

ans = add_numbers("10.5", "20")
print(f"Total is: {ans}")
```

```python
def safe_math(func):
    def wrapper(a, b):
      try :
          return func(a,b)
      except ZeroDivisionError as e :
          print(e)
    return wrapper



@safe_math
def divide_numbers(a, b):
    return a / b

print("Answer 1:", divide_numbers(10, 2))
print("Answer 2:", divide_numbers(10, 0))
```
