```python
# Using list with Exception handling and loops

raw_ages = ["22", "25", "NA", "18", "Error", "30", "Unknown"]
age = []
for i in raw_ages:
   try:
    all_int = int(i)
    age.append(all_int)
   except ValueError:
       pass
print(age)
```

```python
# One more Method : With list comprehension

age = [ i for i in raw_ages if i.isnumeric()]
print(age)
```
