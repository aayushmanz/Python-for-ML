Mission 1: The Birthday Filter (List Comprehension)
Task: You are given a list of random ages.
ages = [15, 22, 18, 25, 22, 19, 30, 21, 22]

Write a single-line list comprehension to create a new list called adults. This new list should only contain ages that are 22 or older.

Hint: Your expression is just the item itself as it loops through the list.

```python
ages = [15, 22, 18, 25, 22, 19, 30, 21, 22]
[adult for adult in ages if adult >= 22]
```

**Mission 2:** Dinosaur Name Cleaner (List Comprehension)
Task: You are given a list of dinosaur names in lowercase.
dinos = ['t-rex', 'raptor', 'spinosaurus', 'brachiosaurus']

Write a single-line list comprehension to create a new list. It must do two things simultaneously:

1. Convert the valid names to Title Case (first letter capitalized).

2. Keep ONLY the names that have a length strictly greater than 6 characters.

3. Expected Output: ['Spinosaurus', 'Brachiosaurus']

```python
dinos = ['t-rex', 'raptor', 'spinosaurus', 'brachiosaurus']
[i.capitalize() for i in dinos if i.lower()]
```

```python
dinos = ['t-rex', 'raptor', 'spinosaurus', 'brachiosaurus']
[i.capitalize() for i in dinos if len(i) > 6]
```

**Mission 3:** The Exam Scorecard (Dictionary Comprehension)
Task: You are given a dictionary containing BCA subjects and their respective marks.
scores = {'Accounting': 45, 'OS': 88, 'Python': 95, 'Maths': 35}

Write a single-line dictionary comprehension to create a new dictionary called passed_subjects. It should only include the subjects where the score is 50 or higher.

Hint: You will need to iterate over scores.items() so you can access both the key (subject) and the value (marks) at the same time.

```python
scores = {'Accounting': 45, 'OS': 88, 'Python': 95, 'Maths': 35}
passed_subjects = {key : value for key,value in scores.items() if value > 50}

print(passed_subjects)
```

**Mission 4:** The ML Feature Mapper (Dictionary Comprehension)
Task: In Machine Learning, you constantly map and transform features. You are given a list of feature names:
features = ['salary', 'age', 'experience', 'location']

Write a single-line dictionary comprehension that generates a dictionary directly from this list.

The Key must be the feature's name.

The Value must be the length (character count) of that specific name.

Expected Output: {'salary': 6, 'age': 3, 'experience': 10, 'location': 8}

```python
features = ['salary', 'age', 'experience', 'location']
dict_feature = {i : len(i) for i in features}
print(dict_feature)
```
