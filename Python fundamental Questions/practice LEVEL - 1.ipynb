1. **Factorial Finder:** _Take an integer n as input and calculate its factorial (e.g., 5! = 120) using a for or while loop._

```python
def factor(n):
    fact = 1
    for i in range(1, n+1):
       fact = fact * i

    print(fact)
```

```python
n = int(input('Enter the Number : '))
factor(n)
```

2. **Prime Number Checker:** _Write a function that accepts a number and checks whether it is a Prime number or not._

**Hint:** _Iterate from 2 to n-1._

```python
def prime(n):
    for i in range(1, n+1):
        if n%i == 0:
            print(f"{n} is not prime number ")
            break
    else :
        print(f"{n} is prime number ")
```

```python
n = int(input('enter the number : '))
prime(n)
```

5. **Second Largest Number:** _Given a list [10, 20, 4, 45, 99]. Find the Second Largest number in the list without using the built-in sort() method or max() function repeatedly._

```python
l = [10, 20, 4, 45, 99]
maximum = max(l)
second_max = 0
for i in l:
    if i > second_max and i < maximum:
           second_max = i
print(second_max)
```

7. Character Frequency Counter: Take a string input (e.g., "programming"). Return a dictionary that contains the count of every character in that string.

Output: {'p': 1, 'r': 2, 'o': 1, 'g': 2, ...}

```python
my = "programming"
count = {}

for i in my:
    if i in count:
        count[i] = count[i] + 1
    else :
        count[i] = 1
print(count)
```

```python
# Example :
my = "programming"
count = {i : my.count(i) for i in my}
print(count)
```

```python
from collections import Counter

my = "programming"

count = dict(Counter(my))

print(count)
```

```python
my = "programming"

count = {my.count("")}
print(count)
```
