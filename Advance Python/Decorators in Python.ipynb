**Decorators**

_A decorator in python is a function that receives another function as input and adds some functionality(decoration) to and it and returns it._

This can happen only because python functions are 1st class citizens.

_There are 2 types of decorators available in python_

* Built in decorators like @staticmethod, @classmethod, @abstractmethod and @property etc
* User defined decorators that we programmers can create according to our needs

```python
from discord.ext.commands.parameters import empty


# Python are 1 st class function

def modify(func, num):
    return func(num)
def square(num):
    return num ** 2

modify(square, 9)
```

```python
# simple example

def my_decor(func):
    def wrapper():
        print("*"*23)
        func()
        print("*"*23)
    return wrapper

def hello():
    print('hello')

def display():
    print('hello ayush')
a = my_decor(hello)
a()

b = my_decor(display)
b()
```

```python
# example code :
def my_decor(func):
    def wrapper():
        print("*"*23)
        func()
        print("*"*23)
    return wrapper
@my_decor
def hello():
    print('hello')
@my_decor
def display():
    print('hello ayush')
hello()
display()
```

```python
# anything meaningful ?

import time

def timer(func):
    def wrapper(*args):
        start = time.time()
        func(*args)
        print('time taken by', func.__name__, time.time()-start,'secs')
    return wrapper
@timer
def hello():
    print('ayush is a male ')
    time.sleep(2)

@timer
def display():
    print('displaying somethings')
    time.sleep(5)
@timer
def square(num):
    time.sleep(2)
    print(num**2)
square(3)
display()
hello()
```

```python
# Examples for Decorators  :

def typechecking(datatype):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if type(func(*args[0:])) == datatype:
                return func(*args[0:])
            else:
                raise TypeError('Bro please only write int value type !')

        return inner_wrapper

    return outer_wrapper


@typechecking(int)
def divide(a, b):
    return a // b


a = int(input("Enter the a value : "))
b = int(input("Enter the b value : "))

print(divide(a, b))
```

```python
# some Questions Solutions :

def log_action(func):
    def wrapper_inner(filename):
            print(f'{func.__name__} is now started !')
            func(filename)
            print(f'{func.__name__} is now Completed !')
    return wrapper_inner


@log_action
def load_data(filename):
    print(f"Loading {filename} into Pandas...")

load_data("users.csv")
```

```python
# More Questions and Solutions:

def positive_only(func):
  def wrapper(*args):
      for i in args:
        if i > 0:
            print('Ok values are checked !')
            return func(*args)
        else:
            raise Exception('Dekho Number toh positive hi chaiye hoga !')
      return None

  return wrapper



@positive_only
def calculate_total_salary(basic, bonus):
    return basic + bonus

print(calculate_total_salary(50000, 10000))
print(calculate_total_salary(50000, -5000))
```
