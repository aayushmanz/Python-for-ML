**Namespaces**

_A namespace is a space that holds names (identifiers). Programmatically speaking, namespaces are dictionary of identifiers (keys) and their objects (values)_

**There are 4 types of namespaces:**

* Builtin Namespace
* Global Namespace
* Enclosing Namespace
* Local Namespace

**Scope and LEGB Rule**

_A scope is a textual region of a Python program where a namespace is directly accessible._

_The interpreter searches for a name from the inside out, looking in the local, enclosing, global, and finally the built-in scope. If the interpreter doesn't find the name in any of these locations, then Python raises a NameError exception._

```python
# local and global

a = 2

def temp():
    b = 3
    print(b)
temp()
print(a)
```

```python
# local and global -> same name
a = 2

def temp():
    a = 3
    print(a)
temp()
print(a)
```

```python
# local and global -> local does not have but global has
a = 2

def temp():
    #local
    print(a) # local can take the global value !
temp()
print(a)
```

```python
# local and global -> editing global
a = 2

def temp():
  a += 1      # we can not change the global variable in local.
  print(a)

temp()
print(a)
```

```python
a = 2

def temp():
    # local var
    global a       # becuz of global keyword we can change local variable
    a += 1
    print(a)
temp()
print(a)
```

```python
# local and global -> global created inside local
a = 2

def temp():
    # local var
    global a
    a = 1
    print(a)
temp()
print(a)
```

```python
# local and global -> function parameter is local

def temp(z):
    # local var
    print(z)
a = 5
temp(5)
print(a)
```

```python
# built-in scope
print('hello') # built-in scope

import builtins   # built - in imported
print(dir(builtins))
```

```python
# renaming built-ins
l = [1,2,3] # global scope
max(l) # built - in scope

def max():    # we can not access the built-in function in local scope !
    print("hello")

max(l)
```

```python
# enclosing scope

def outer():  # Enclosing scope
    a = 3
    def inner(): # local scope
        print(a)
        print("inner function")
    inner()
    print('outer function')

a = 1 # global scope
outer()
print("main program")
```

```python
# nonlocal keyword :

def outer():  # Enclosing scope
    a = 3
    def inner(): # local scope can not change enclosing scope
        nonlocal a # becuz of this we can make changes !
        a += 1
        print("inner function", a)
    inner()
    print('outer function' , a)

outer()
print("main program")
```
